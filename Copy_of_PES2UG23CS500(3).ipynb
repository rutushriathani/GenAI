```python
from transformers import pipeline

# NER Pipeline: Identifies names, organizations, and locations
# Model 'dslim/bert-base-NER' is a standard bert-base model fine-tuned for NER
ner_analyzer = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")

# Sentiment Pipeline: Classifies emotion as Positive or Negative
# Model 'bert-base-uncased' fine-tuned for sentiment
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
def analyze_feedback(text):
    # 1. Detect entities (Who is being talked about?)
    entities = ner_analyzer(text)

    # 2. Detect sentiment (Is it happy or angry?)
    sentiment = sentiment_analyzer(text)[0]

    return {
        "text": text,
        "entities": [(e['word'], e['entity_group']) for e in entities],
        "sentiment": sentiment['label'],
        "confidence": f"{sentiment['score']:.2%}"
    }
```

```python
review = "I bought the new iPhone from the Apple Store in New York, but I am angry because the battery life is terrible."

result = analyze_feedback(review)

print(f"Entities Found: {result['entities']}") # Output: Apple (ORG), New York (LOC)
print(f"Final Sentiment: {result['sentiment']}") # Output: NEGATIVE
```
