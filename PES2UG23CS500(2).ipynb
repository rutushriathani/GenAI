```python
from transformers import pipeline

bert_mlm = pipeline("fill-mask", model="bert-base-uncased")
roberta_mlm = pipeline("fill-mask", model="roberta-base")
bart_gen = pipeline("text-generation", model="facebook/bart-base")
```

```python
bart_gen("The future of Artificial Intelligence is", max_length=50)
```

```python
bert_mlm("The goal of Generative AI is to [MASK] new content.")
roberta_mlm("The goal of Generative AI is to <mask> new content.")
```

```python
qa_base = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

qa_base(
    question="What are the risks?",
    context="Generative AI poses risks such as hallucinations, bias, and deepfakes."
)
```

```python
from transformers import pipeline

models = ["bert-base-uncased", "roberta-base", "facebook/bart-base"]
prompt = "The future of Artificial Intelligence is"

print("--- Experiment 1: Text Generation ---")
for model_name in models:
    try:
        # Use the text-generation pipeline
        generator = pipeline("text-generation", model=model_name)
        result = generator(prompt, max_new_tokens=10, do_sample=True)
        print(f"Model: {model_name} -> Output: {result[0]['generated_text']}")
    except Exception as e:
        print(f"Model: {model_name} -> FAILED: {str(e)}")
```
